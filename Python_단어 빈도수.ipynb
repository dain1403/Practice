```python
f = open('C:/Users/t4230/DeskTop/fashion2.txt', 'r', encoding='UTF8')
```

```python
lines = f.read()
```

```python
from konlpy.tag import Twitter

nlpy = Twitter()

nouns = nlpy.nouns(lines)

print(nouns)
```

```python
from collections import Counter

count = Counter(nouns)

tag_count = []

tags = []
```

```python
count
```

```python
for n, c in count.most_common(1000):       ###  '.most_common' :  빈도수가 높은 n개의 명사만 추출
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 99:      ## 2글자 이상 ,  len(tags) <= 49 : 리스트추가개수?
        tag_count.append(dics)
        tags.append(dics['tag'])  ## 키워드만 활용하기 위해 따로 'append'
```

```python
tag_count
```

```python
tags
```

```python
len(tags)
```

```python
for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
```
